```python
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('/content/drive/MyDrive/TCV Lab/Images/hamster.jpg', 1)

b, g, r = cv2.split(img)

plt.imshow(b, cmap='gray')
plt.title('Blue Channel (from spilt)')
plt.axis('off')
plt.show()
```

```python
import cv2
from matplotlib import pyplot as plt

img = cv2.imread('/content/drive/MyDrive/TCV Lab/Images/hamster.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

(h, w, c) = img.shape # Define h and w from img.shape
center = (w // 2, h // 2)

M = cv2.getRotationMatrix2D(center, -45, 1.0)
rotated = cv2.warpAffine(img, M, (w, h))

plt.imshow(rotated)
plt.title('Rotated Image') # Corrected: plt.axis to plt.title
plt.axis('off')
plt.show()
```

```python
import cv2
from matplotlib import pyplot as plt

%matplotlib inline

img = cv2.imread('/content/drive/MyDrive/TCV Lab/Images/mmu.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
logo = cv2.imread('/content/drive/MyDrive/TCV Lab/Images/logo.png')
logo = cv2.cvtColor(logo, cv2.COLOR_BGR2RGB)

# We want to put logo on top-left corner, so we create a ROI
(rows, cols, channels) = logo.shape
roi = img[0:rows, 0:cols]
plt.imshow(roi)
plt.axis('off')
```

```python
img2gray = cv2.cvtColor(logo, cv2.COLOR_BGR2GRAY)
(ret, mask) = cv2.threshold(img2gray, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)

plt.imshow(mask, cmap='gray')
plt.axis('off')
```

```python
mask_inv = cv2.bitwise_not(mask)
plt.imshow(mask_inv, cmap='gray')
plt.axis('off')
```

```python
logo_fg = cv2.bitwise_and(roi,roi,mask = mask_inv)
plt.imshow(logo_fg)
plt.axis('off')
```

```python
logo_fg = cv2.bitwise_and(logo,logo,mask = mask_inv)
plt.imshow(logo_fg)
plt.axis('off')
```

```python
logo_bg = cv2.bitwise_and(roi, roi, mask = mask)
# Put logo in ROI
logo_ROI = cv2.add(logo_bg, logo_fg)
plt.imshow(logo_ROI)
plt.axis('off')
```

```python
import matplotlib.pyplot as plt

# Put the ROI with logo in the image
img[0:rows, 0:cols] = logo_ROI
plt.imshow(img)
plt.axis('off')
```
